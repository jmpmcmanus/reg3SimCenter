```python
# Import python modules
import pandas as pd
import geopandas as gpd
import netCDF4 as nc
import numpy as np
from scipy.interpolate import griddata
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from shapely.geometry import Polygon
```

```python
# Read netCDF file, and display contents
filename = 'var_dp3r3b1c1h1l1_swan63_subset.nc'
ds = nc.Dataset(filename, 'r')
ds.variables
```

```python
# Extract variables
depth = ds.variables['depth'][:].data
hs = ds.variables['hs'][:].data
tps = ds.variables['tps'][:].data
wavedir = ds.variables['dir'][:].data
lon = ds.variables['x'][:].data
lat = ds.variables['y'][:].data
```

```python
# Define lowe left and upper right longitude and latitude coordinates
lllon = min(lon)
lllat = min(lat)
urlon = max(lon)
urlat = max(lat)

# set up basemap chose projection!
m = Basemap(projection = 'merc', resolution='f',llcrnrlon = lllon, llcrnrlat = lllat, urcrnrlon = urlon, urcrnrlat = urlat)
m_lon, m_lat = m(*(lon, lat))
```

```python
# generate grid data
numcols, numrows = 960, 960
xi = np.linspace(m_lon.min(), m_lon.max(), numcols)
yi = np.linspace(m_lat.min(), m_lat.max(), numrows)
xi, yi = np.meshgrid(xi, yi)

# interpolate, there are better methods, especially if you have many datapoints
zid = griddata((m_lon, m_lat), depth, (xi, yi), method='linear', fill_value=np.nan, rescale=False)
zih = griddata((m_lon, m_lat), hs[3], (xi, yi), method='linear', fill_value=np.nan, rescale=False)
zit = griddata((m_lon, m_lat), tps[3], (xi, yi), method='linear', fill_value=np.nan, rescale=False)
zir = griddata((m_lon, m_lat), wavedir[3], (xi, yi), method='linear', fill_value=np.nan, rescale=False)
```

```python
# define fig and ax
fig, ax = plt.subplots(figsize=(12, 10))

# draw map details
m.drawmapboundary(fill_color = 'olive', zorder = 1)

# Plot interpolated depth data
m.contourf(xi, yi, zid, 500, cmap='magma', zorder = 2)
cbar = plt.colorbar()
plt.title('Depth')
plt.show()
```

```python
# define fig and ax
fig, ax = plt.subplots(figsize=(12, 10))

# draw map details
m.drawmapboundary(fill_color = 'olive', zorder = 1)

# Plot interpolated hs data
m.contourf(xi, yi, zih, 500, cmap='magma', zorder = 2)
cbar = plt.colorbar()
plt.title('HS')
plt.show()
```

```python
# define fig and ax
fig, ax = plt.subplots(figsize=(12, 10))

# draw map details
m.drawmapboundary(fill_color = 'olive', zorder = 1)

# Plot interpolated tps data
m.contourf(xi, yi, zit, 500, cmap='magma', zorder = 2)
cbar = plt.colorbar()
plt.title('TPS')
plt.show()
```

```python
# define fig and ax
fig, ax = plt.subplots(figsize=(12, 10))

# draw map details
m.drawmapboundary(fill_color = 'olive', zorder = 1)

# Plot interpolated dir data
m.contourf(xi, yi, zir, 500, cmap='magma', zorder = 2)
cbar = plt.colorbar()
plt.title('DIR')
plt.show()
```
